```python
from tkinter import *
from tkinter import messagebox

root = Tk()
root.geometry("1200x500")
root.title("Bill Management")
root.resizable(False, False)

def Reset():
    entry1.delete(0,END)
    entry2.delete(0,END)
    entry3.delete(0,END)
    entry4.delete(0,END)
    entry5.delete(0,END)
    entry6.delete(0,END)
    entry7.delete(0,END)
    
def Total():
        try:
            a1 = int(maggie.get())
        except: 
            a1=0
        try:
            a2 = int(dosa.get())
        except:
            a2=0
        try:  
            a3 = int(manchurian.get())
        except:
            a3=0
        try:
            a4 = int(fries.get())
        except:
            a4=0
        try:
            a5 = int(idli.get())
        except:
            a5=0
        try:
            a6 = int(tea.get())
        except:
            a6=0
        try:
            a7 = int(coffee.get())
        except:
            a7=0
            
        #defining cost of each item/quantity
        cost1=60*a1
        cost2=100*a2
        cost3=80*a3
        cost4=90*a4
        cost5=95*a5
        cost6=20*a6
        cost7=30*a7
        
        lbl_total=Label(f2,font=('aria',20,'bold'),text="Total",width=18,fg="darkseagreen2",bg="black")
        lbl_total.place(x=0,y=50)
    
        entry_total=Entry(f2,font=('aria',20,'bold'),textvariable=Total_bill,bd=6,width=15,bg="darkseagreen2")
        entry_total.place(x=25,y=100)
        
        total_cost=cost1+cost2+cost3+cost4+cost5+cost6+cost7
        string_bill="Rs.",str('%.2f' %total_cost)
        Total_bill.set(string_bill)
        
Label(text="BILL MANAGEMENT",bg="darkslategray",fg="darkolivegreen1",font=("calibri",33),width="300",height="2").pack()

#frame for menu card
f=Frame(root,bg="darkseagreen2", highlightbackground="black",highlightthickness=1,width=300,height=370)
f.place(x=20,y=118)

Label(f,text="Menu", font=("Gabriola",40,"bold"),fg="Black",bg="darkseagreen2").place(x=80,y=0)

Label(f,font=("Lucida Calligraphy", 13, 'bold'),text="Maggie........Rs.60/plate",fg="black",bg="darkseagreen2").place(x=10,y=80)
Label(f,font=("Lucida Calligraphy", 13, 'bold'),text="Dosa........Rs.100/plate",fg="black",bg="darkseagreen2").place(x=10,y=110)
Label(f,font=("Lucida Calligraphy", 13, 'bold'),text="Manchurian......Rs.80/plate",fg="black",bg="darkseagreen2").place(x=10,y=140)
Label(f,font=("Lucida Calligraphy", 13, 'bold'),text="Fries........Rs.90/plate",fg="black",bg="darkseagreen2").place(x=10,y=170)
Label(f,font=("Lucida Calligraphy", 13, 'bold'),text="Idli ........Rs.95/plate",fg="black",bg="darkseagreen2").place(x=10,y=200)
Label(f,font=("Lucida Calligraphy", 13, 'bold'),text="Tea........Rs.20/cup",fg="black",bg="darkseagreen2").place(x=10,y=230)
Label(f,font=("Lucida Calligraphy", 13, 'bold'),text="Coffee........Rs.30/cup",fg="black",bg="darkseagreen2").place(x=10,y=260)

#frame for Bill
f2=Frame(root,bg="darkseagreen2",highlightbackground="black",highlightthickness=1, height=370,width=300,relief=RAISED)
f2.pack()
f2.place(x=750,y=118)

bill=Label(f2,text="Bill",font=('calibri',20),bg="darkseagreen2")
bill.place(x=120, y=10)

#frame For Entry 
f1=Frame(root,bd=5,highlightbackground="black",highlightthickness=1,height=370,width=300,relief=RAISED)
f1.pack()
f1.place(x=350,y=118)

maggie=StringVar()
dosa=StringVar()
manchurian=StringVar()
fries=StringVar()
idli=StringVar()
tea=StringVar()
coffee=StringVar()
Total_bill=StringVar()

#label for maggie
label_maggie=Label(f1,font=("aria",20,'bold'),text="Maggie",width=12,fg="darkslategray")
label_maggie.grid(row=1,column=0)

#label for dosa
label_dosa=Label(f1,font=("aria",20,'bold'),text="Dosa",width=12,fg="darkslategray")
label_dosa.grid(row=2,column=0)

#label for machurian
label_manchurian=Label(f1,font=("aria",20,'bold'),text="Manchurain",width=12,fg="darkslategray")
label_manchurian.grid(row=3,column=0)

#label for fries
label_fries=Label(f1,font=("aria",20,'bold'),text="Fries",width=12,fg="darkslategray")
label_fries.grid(row=4,column=0)

#label for Idli Sambhar
label_idli=Label(f1,font=("aria",20,'bold'),text="Idli Sambhar",width=12,fg="darkslategray")
label_idli.grid(row=5,column=0)

#label for tea
label_tea=Label(f1,font=("aria",20,'bold'),text="Tea",width=12,fg="darkslategray")
label_tea.grid(row=6,column=0)

#label for coffee
label_coffee=Label(f1,font=("aria",20,'bold'),text="Coffee",width=12,fg="darkslategray")
label_coffee.grid(row=7,column=0)

#Entry for maggie
entry1=Entry(f1,font=("aria",20,'bold'),textvariable=maggie,bd=6,width=8,bg="darkseagreen3")
entry1.grid(row=1,column=1)

#Entry for dosa
entry2=Entry(f1,font=("aria",20,'bold'),textvariable=dosa,bd=6,width=8,bg="darkseagreen3")
entry2.grid(row=2,column=1)

#Entry for manchurian
entry3=Entry(f1,font=("aria",20,'bold'),textvariable=manchurian,bd=6,width=8,bg="darkseagreen3")
entry3.grid(row=3,column=1)

#entry for fries
entry4=Entry(f1,font=("aria",20,'bold'),textvariable=fries,bd=6,width=8,bg="darkseagreen3")
entry4.grid(row=4,column=1)

#Entry for idli sambhar
entry5=Entry(f1,font=("aria",20,'bold'),textvariable=idli,bd=6,width=8,bg="darkseagreen3")
entry5.grid(row=5,column=1)

#Entry for tea
entry6=Entry(f1,font=("aria",20,'bold'),textvariable=tea,bd=6,width=8,bg="darkseagreen3")
entry6.grid(row=6,column=1)

#Entry for coffee
entry7=Entry(f1,font=("aria",20,'bold'),textvariable=coffee,bd=6,width=8,bg="darkseagreen3")
entry7.grid(row=7,column=1)

#for button
btn=Button(f1,bd=5,fg="black",bg="darkseagreen4",font=("ariel",16,"bold"),width=10,text="Reset",command=Reset)
btn.grid(row=8,column=0)

btn_total=Button(f1,bd=5,fg="black",bg="darkseagreen4",font=("ariel",16,"bold"),width=10,text="Total",command=Total)
btn_total.grid(row=8,column=1)


root.mainloop()
```
